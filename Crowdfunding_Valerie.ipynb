### Section 9-Map
    - Use map() to transform some of the data

```python
# Use mapping to transform some data of the csv

def map_goal_to_value(Goal):
    if Goal < 7500:
        return 'Low'
    elif 7500 <= Goal < 12000:
        return 'Medium'
    else:
        return 'High'
```

```python
# Apply the function to create a new category
# I'm using the filtered DataFrame for this

Kickstarter['Goal_Value'] = Kickstarter['Goal'].apply(map_goal_to_value)
```

```python
# Display the modified DataFrame with the new column 'Value'

Kickstarter.head(5)
```
